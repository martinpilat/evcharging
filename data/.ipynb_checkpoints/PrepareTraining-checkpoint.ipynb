```python
import pandas as pd
import datetime

houses = pd.read_csv('selected_1.csv', parse_dates=[0])
houses.fillna(method='ffill', inplace=True)

baseline = pd.read_csv('baseline_1.csv', parse_dates=[0], header=None)
baseline.columns=['datetime', 'consumption']
baseline.fillna(method='ffill', inplace=True)

requests = pd.read_csv('requests_1.csv', parse_dates=[4,5])
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2013,4,1)

f_baseline = baseline[(baseline['datetime'] >= start) & (baseline['datetime']<end)]
f_houses = houses[(houses['datetime'] >= start) & (houses['datetime']<end)]

f_baseline.set_index('datetime',inplace=True)
f_houses.set_index('datetime',inplace=True)

uncontrolable = f_houses.sum(axis=1) + 50/450*f_baseline.sum(axis=1)
```

```python
from collections import namedtuple

Request = namedtuple('Request', ['houseid','available_after', 'available_before', 'missing_charge', 'charging_speed'])

reqs = []

for r in range(len(requests)):
    l = requests.iloc[r]
    reqs.append(Request(houseid=l['houseid'], available_after=l['available_after'], available_before=l['available_before'], 
                        missing_charge=l['required_charge'], charging_speed=l['max_charging_speed']))
```

# The optimal charging, minimizing directly the variance of consumption

```python
from collections import defaultdict

import cvxpy as cp

start = datetime.datetime(2013,1,1)
end = datetime.datetime(2013,4,1)

vars_time = defaultdict(list)

constraints = []

sum_charges = 0
for r in reqs:
    if (r.available_after < start) or (r.available_before > end):
        continue
    cur_time = r.available_after
    if cur_time < start:
        print(r)
    sum_charges += r.missing_charge
    var = cp.Variable(shape=(r.available_before - cur_time)//datetime.timedelta(minutes=30))
    constraints.append(0.0 <= var)
    constraints.append(var <= r.charging_speed)
    constraints.append(cp.sum(var)/2 == r.missing_charge)
    i = 0
    while cur_time < r.available_before:
        vars_time[cur_time].append((r.houseid,var[i]))
        cur_time += datetime.timedelta(minutes=30)
        i += 1

N = len(uncontrolable)
    
var_e2x = ((sum(uncontrolable) + 2*sum_charges)/N)**2
var_ex2 = 0

for t in uncontrolable.index:
    if t not in vars_time.keys():
        var_ex2 += uncontrolable[t]**2/N
    
for t,v in vars_time.items():
    var_ex2 += cp.square(uncontrolable[t] + cp.sum([x[1] for x in v]))/N

obj_expr = var_ex2 - var_e2x
    
obj = cp.Minimize(obj_expr)

prob = cp.Problem(obj, constraints)
prob.solve(verbose=True)  # Returns the optimal value.
print("status:", prob.status)
print("optimal value", prob.value)  

charging = pd.Series()

for k,v in vars_time.items():
    values = [x[1].value for x in v]
    charging[k] = sum(values)  

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
charging = charging.reindex(uncontrolable.index).fillna(0.0)
print(np.std((uncontrolable + charging)))
plt.figure(figsize=(20,5))
plt.plot(uncontrolable)
plt.plot(uncontrolable + charging)
print(f'{sum_charges} == {sum(charging)/2}')

optimum_charge = f_houses.copy()

for c in optimum_charge.columns:
    optimum_charge[c].values[:] = 0

for k, v in vars_time.items():
    for var in v:
        optimum_charge[var[0]][k] = var[1].value
        
optimum_charge.to_csv('optimum_charge.csv')
```

# Minimizing the variance of each household separately

```python
from collections import defaultdict

import cvxpy as cp

vars_time = defaultdict(list)

constraints = []

start = datetime.datetime(2013, 1, 1)
end = datetime.datetime(2013, 4, 1)

var_ex2 = 0.0

req_lens = defaultdict(int)

for r in reqs:
    req_lens[r.houseid] += (r.available_before - r.available_after)//datetime.timedelta(minutes=30)

sum_charges = 0
N = 0
for r in reqs:
    if (r.available_after < start) or (r.available_before > end):
        continue
    cur_time = r.available_after
    if cur_time < start:
        print(r)
    sum_charges += r.missing_charge
    var = cp.Variable(shape=(r.available_before - cur_time)//datetime.timedelta(minutes=30))
    constraints.append(0.0 <= var)
    constraints.append(var <= r.charging_speed)
    constraints.append(cp.sum(var)/2 == r.missing_charge)
    var_ex2 += cp.sum(cp.square(var + f_houses[r.houseid][r.available_after:r.available_before - datetime.timedelta(minutes=30)].values))/req_lens[r.houseid]/1000
    i = 0
    while cur_time < r.available_before:
        vars_time[cur_time].append((r.houseid,var[i]))
        cur_time += datetime.timedelta(minutes=30)
        N += 1
        i += 1

obj_expr = var_ex2
    
obj = cp.Minimize(obj_expr)

prob = cp.Problem(obj, constraints)
prob.solve(verbose=True)  # Returns the optimal value.
print("status:", prob.status)
print("optimal value", prob.value)  

charging = pd.Series()

for k,v in vars_time.items():
    values = [x[1].value for x in v]
    charging[k] = sum(values)  

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
charging = charging.reindex(uncontrolable.index).fillna(0.0)
print(np.std((f_houses.sum(axis=1) + charging)))
plt.figure(figsize=(20,5))
plt.plot(f_houses.sum(axis=1))
plt.plot(f_houses.sum(axis=1) + charging)
print(f'{sum_charges} == {sum(charging)/2}')

optimum_charge = f_houses.copy()

for c in optimum_charge.columns:
    optimum_charge[c].values[:] = 0

for k, v in vars_time.items():
    for var in v:
        optimum_charge[var[0]][k] = var[1].value
        
optimum_charge.to_csv('optimum_charge_hh.csv')
```

# Minimizing the global variance and the changes during charging

```python
from collections import defaultdict

import cvxpy as cp

vars_time = defaultdict(list)

constraints = []

sum_charges = 0
N = 0
stable_obj = 0.0
for r in reqs:
    if (r.available_after < start) or (r.available_before > end):
        continue
    cur_time = r.available_after
    if cur_time < start:
        print(r)
    sum_charges += r.missing_charge
    req_len = (r.available_before - cur_time)//datetime.timedelta(minutes=30)
    var = cp.Variable(shape=req_len)
    constraints.append(0.0 <= var)
    constraints.append(var <= r.charging_speed)
    constraints.append(cp.sum(var)/2 == r.missing_charge)
    stable_obj += cp.sum(cp.square(cp.diff(var)))/(req_len - 1)
    i = 0
    while cur_time < r.available_before:
        vars_time[cur_time].append((r.houseid,var[i]))
        cur_time += datetime.timedelta(minutes=30)
        N += 1
        i += 1

N = len(uncontrolable)
    
var_e2x = ((sum(uncontrolable) + 2*sum_charges)/N)**2
var_ex2 = 0

for t in uncontrolable.index:
    if t not in vars_time.keys():
        var_ex2 += uncontrolable[t]**2
    
for t,v in vars_time.items():
    var_ex2 += cp.square(uncontrolable[t] + cp.sum([x[1] for x in v]))

obj_expr = var_ex2/N - var_e2x

#this is where you can change the lambda for the regularization
obj = cp.Minimize(obj_expr + 0.01*stable_obj)

prob = cp.Problem(obj, constraints)
prob.solve(verbose=True)  # Returns the optimal value.
print("status:", prob.status)
print("optimal value", prob.value)  

charging = pd.Series()

for k,v in vars_time.items():
    values = [x[1].value for x in v]
    charging[k] = sum(values)  

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
charging = charging.reindex(uncontrolable.index).fillna(0.0)
print(np.std((uncontrolable + charging)))
plt.figure(figsize=(20,5))
plt.plot(uncontrolable)
plt.plot(uncontrolable + charging)
print(f'{sum_charges} == {sum(charging)/2}')

optimum_charge = f_houses.copy()

for c in optimum_charge.columns:
    optimum_charge[c].values[:] = 0

for k, v in vars_time.items():
    for var in v:
        optimum_charge[var[0]][k] = var[1].value
        
optimum_charge.to_csv('optimum_charge_stable01.csv')
```
